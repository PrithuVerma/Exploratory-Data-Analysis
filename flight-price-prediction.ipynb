```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_excel('/kaggle/input/flight-price/flight_price.xlsx')
df.shape
```

```python
df.info()
```

```python
df['Date'] = df['Date_of_Journey'].str.split('/').str[0]
df['Month'] = df['Date_of_Journey'].str.split('/').str[1]
df['Year'] = df['Date_of_Journey'].str.split('/').str[2]

df['Date'] = df['Date'].astype(int)
df['Month'] = df['Month'].astype(int)
df['Year'] = df['Year'].astype(int)

df.drop('Date_of_Journey',inplace = True,axis = 1)
```

```python
df['Arrival_Time'] = df['Arrival_Time'].apply(lambda x : x.split(' ')[0])
df['Arrival_Hour'] = df['Arrival_Time'].str.split(':').str[0]
df['Arrival_Min'] = df['Arrival_Time'].str.split(':').str[1]

df['Arrival_Hour'].astype(int)
df['Arrival_Min'].astype(int)

df.drop(columns = 'Arrival_Time',inplace = True,axis = 1)
df.head()
```

```python
# df['Dep_Hour'] = df['Dep_Time'].str.split(':').str[0]
# df['Dep_Min'] = df['Dep_Time'].str.split(':').str[1]

# df['Dep_Hour'].astype(int)
# df['Dep_Min'].astype(int)

df['Arrival_Hour'] = df['Arrival_Hour'].astype(int)
df['Arrival_Min'] = df['Arrival_Min'].astype(int)

df['Dep_Hour'] = df['Dep_Hour'].astype(int)
df['Dep_Min'] = df['Dep_Min'].astype(int)

df.head()
df.info()
```

```python
df['Total_Stops'] = df['Total_Stops'].map({'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4,np.nan:1})
df.head(10)
```

```python
df.drop('Route',inplace = True,axis = 1)
df.head()
```

```python
df['Duration_hrs'] = df['Duration'].str.split(' ').str[0].str.split('h').str[0]
df['Duration_min'] = df['Duration'].str.split(' ').str[1].str.split('m').str[1]
df['Duration_min'].replace('NaN',0)

df.head()
```
