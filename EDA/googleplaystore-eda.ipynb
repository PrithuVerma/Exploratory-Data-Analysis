```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

# **Reading the Data**

```python
df = pd.read_csv('/kaggle/input/ggl-ps/googleplaystore.csv')
df.head()
```

# **Checking the Missing Values**

```python
df.isnull().sum()
```

# **Data Cleaning**

```python
df['Reviews'].apply(lambda x : type(x))
```

```python
df['Reviews'].str.isnumeric().sum()
```

**Out of 10841 entries only 1 is in int datatype, to find the below command is used where '~' represents nt or opposite of the statement**


```python
df[~df['Reviews'].str.isnumeric()]
```

```python
df_copy = df.copy()
```

```python
df_copy = df_copy.drop(df_copy.index[10472])
```

```python
df_copy[~df_copy['Reviews'].str.isnumeric()]
```

```python
df_copy['Reviews'] = df_copy['Reviews'].astype(int)
```

```python
df_copy['Reviews'].apply(lambda x : type(x))
```

```python
df_copy['Size'].apply(lambda x : type(x))
```

```python
df_copy['Size'].str.isnumeric().sum()
```

```python
df_copy.Size.unique()
```

```python
df_copy['Size'] = df_copy['Size'].str.replace('M','000')
df_copy['Size'] = df_copy['Size'].str.replace('k','')
df_copy['Size'] = df_copy['Size'].replace("Varies with device", np.nan)
df_copy['Size'].astype(float)
```

```python
df_copy[['Installs','Price']].value_counts()
```

```python
chars = ['+',',','$']
clean_cols = ['Installs','Price']

for items in chars:
    for cols in clean_cols:
        df_copy[cols] = df_copy[cols].str.replace(items,'')
```

```python
df_copy['Installs'] = df_copy['Installs'].astype(int)
df_copy['Price'] = df_copy['Price'].astype(float)
```

```python
df_copy.info()
```

```python
df_copy['Last Updated'] = pd.to_datetime(df_copy['Last Updated'])
df_copy['Date'] = df_copy['Last Updated'].dt.day
df_copy['Month'] = df_copy['Last Updated'].dt.month
df_copy['Year'] = df_copy['Last Updated'].dt.year
```

```python
df_copy.info()
```

```python
df_copy.head()
```

# **EDA**

```python
df_copy[df_copy.duplicated('App')]
```

```python
df_copy.drop_duplicates(subset = ['App'],keep = 'first',inplace = True)
```

```python
df_copy.shape
```

```python
# df_copy['Size'] = df_copy['Size'].astype(float)
numerical_features = [feature for feature in df_copy.columns if df_copy[feature].dtype !='O']
categorical_features = [feature for feature in df_copy.columns if df_copy[feature].dtype =='O']

print('We have {} numerical features : {}'.format(len(numerical_features),numerical_features))
print('We have {} categorical features : {}'.format(len(categorical_features),categorical_features))
```

```python
for col in categorical_features:
    print(df[col].value_counts(normalize = True)*100)
    print('>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>')
```

```python
plt.figure(figsize = (15,15))
plt.suptitle('Univariate Analysis of Numerical Features',fontsize = 20,fontweight = "bold",alpha = 0.8,y = 1.)

for i in range(0,len(numerical_features)):
    plt.subplot(5,3,i+1)
    sns.kdeplot(x=df_copy[numerical_features[i]],shade = True,color = 'r')
    plt.xlabel(numerical_features[i])
    plt.tight_layout()
    
```
