```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/content/Books_Data_Clean.csv')
data.head()
```

```python
data.columns
```

```python
data.describe()
```

```python
data.info()
```

```python
data = data[data['Publishing Year']>1900]
```

```python
data.isna().sum()
```

```python
data.dropna(subset = 'Book Name', inplace = True)
```

```python
data.duplicated().sum()
```

```python
data.nunique()
```

```python
plt.hist(data['Publishing Year'])
plt.xlabel('Publishing Year')
plt.ylabel('Frequency')
plt.title('Distribution of Publishing Year')
plt.show()
```

```python
data['genre'].value_counts().plot(kind = 'bar')
plt.xlabel('Genre')
plt.ylabel('No. of Books')
plt.title('No. of Books in each Genre')
plt.show()
```

```python
data.groupby('Author')['Book_average_rating'].mean()
```

```python
sns.boxplot(x = 'genre', y = 'Book_ratings_count', data = data)
plt.xlabel('Genre')
plt.ylabel('Book Ratings count')
plt.title('Boxplot Ratings in each Genre')
plt.show()
```

```python
plt.scatter(data['sale price'],data['units sold'])
plt.xlabel('Sale Price')
plt.ylabel('Units Sold')
plt.title('Scatter Plot of Sale Price vs Units Sold')
plt.show()
```

```python
languages = data.language_code.value_counts()
```

```python
plt.pie(languages, labels = languages.index, autopct = '%1.1f%%',startangle=90)
plt.title('Distribution of Languages')
plt.show()
```

```python
data.groupby("Publisher ")["publisher revenue"].sum().sort_values(ascending = False)
```

```python
data.groupby('Author_Rating')['Book_ratings_count'].mean().sort_values(ascending = False)
```

```python
data.columns
```

```python
data.groupby('language_code')['Book Name'].count().sort_values(ascending = False)
#-------------OR-------------#
'''data.groupby('language_code').size().sort_values(ascending = False)'''
```

```python
total_sales = data.groupby('Author')['sale price'].sum().sort_values(ascending = False)
total_sales.head(10).plot(kind = 'bar')
plt.xlabel('Author')
plt.ylabel('Total Sales')
plt.title('Top 10 Authors by Total Sales')
```

```python
sns.boxplot(x = 'Author_Rating', y = 'units sold', data = data)
plt.xlabel('Author Rating')
plt.ylabel('Units Sold')
plt.title('Boxplot of Author Rating')
```

```python
data.groupby('Publishing Year')['units sold'].sum().plot(kind = 'line')
plt.xlabel('Publishing Year')
plt.ylabel('Units Sold')
plt.title('Units Sold Over the Years')
```
